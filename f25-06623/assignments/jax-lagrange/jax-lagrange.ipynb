A prototypical example of a constrained optimization (from Greenberg, Advanced Engineering Mathematics, Ch 13.7) is to find the point on a plane that is closest to the origin. The plane is defined by the equation 2𝑥 − 𝑦 + 𝑧 = 3, and we seek to minimize $𝑥^2 + 𝑦^2 + 𝑧^2$ (the squared distance from the origin) subject to the equality constraint defined by the plane.

First, solve this problem with `scipy.optimize.minimize` and constraints.



This is classically solved with Lagrangian optimization. We set up an augmented problem:

𝐹(𝑋) = 𝑓(𝑋) − 𝜆𝑔(𝑋) where 𝑓(𝑋) is the original function to minimize, and 𝑔(𝑋) is the constraint. then solve the simultaneous equations:

d𝐹/d𝑥 (𝑋) = d𝐹/d𝑦 (𝑋) = d𝐹d𝑧 (𝑋) = 𝑔(𝑋) = 0
