# 📘 Optimizing Your Daily Routine for Academic Success

You’re a student working to design your ideal daily schedule. Every day, you have 16 hours to divide among three essential activities:

🎓 1. Studying

Studying directly improves your academic performance. The benefit follows a law of diminishing returns: your first few study hours are very productive, but each additional hour helps a bit less.

✅ Effect on Academic Performance: $+10 \cdot \sqrt{s}$

⸻

🧑‍🤝‍🧑 2. Socializing

Socializing has a mixed impact:

* 	Up to a point, it boosts your academic performance — by improving mental health and reducing burnout.
* 	Beyond that point, it becomes a distraction and reduces your academic focus.

We model this as an inverted parabola:
$\text{Socializing Contribution} = a \cdot z - b \cdot z^2$
Where:

* 	$z$ is hours of socializing
* 	$a$ = 6 (initial benefit)
* 	$b$ = 0.75 (rate at which it turns negative)

⸻

💆 3. Self-care

Self-care (sleep, meals, exercise, downtime) doesn’t directly raise your academic score, but it’s essential for physical and mental health.
You must spend at least 6 hours on self-care each day.

⏳ Constraint: $c \geq 6$

⸻

🎯 Your Goal

Maximize your Academic Performance Score, defined as:

$\text{Score}(s, z, c) = 10 \cdot \sqrt{s} + 6z - 0.75z^2$

Solve this time allocation problem.
